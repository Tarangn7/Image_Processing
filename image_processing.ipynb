## image conversion

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
plt.axis("off")
plt.show()
```

```python
img_gray.shape
```

```python
# Convert BGR to RGB
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_RGB)
plt.axis("off")
```

```python
#Hue, Sturation, Value(Intensity)
img_HSV = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)

plt.imshow(img_HSV)
plt.axis("off")
plt.show()
```

```python
img_g = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2Lab)

plt.imshow(img_g)
plt.axis("off")
plt.show()
```

```python
cv2.imwrite("tommy.jpg",img_gray)
```

```python
blue = img[:,:,0]
green = img[:,:,1]
red = img[:,:,2]

new = np.hstack((blue,green,red))
```

### flipping

```python
# vertical flip
imv = cv2.flip(img,0)
```

```python
plt.imshow(imv)
plt.axis("off")
plt.show()
```

```python
# horizontal flip

imh = cv2.flip(img,1)
plt.imshow(imh)
plt.axis("off")
plt.show()
```

```python
imc = cv2.flip(img,-1)
plt.imshow(imc)
plt.axis("off")
plt.show()
```

### Drawing stuff

```python
# creating a black image
im = np.zeros((512,512,3))
plt.imshow(im)
plt.axis("off")
plt.show()
```

```python
#rectangle
cv2.rectangle(im, pt1 = (100,100),pt2 = (300,300),color = (255,0,0),thickness = -1)
#circle
cv2.circle(im, center = (100,400),radius = 60,color = (0,255,0),thickness = 2)
#line
cv2.line(im, pt1 = (0,0), pt2 = (512,512), thickness = 2, color = (0,0,255))

plt.imshow(im)
plt.axis("off")
plt.show()
```
